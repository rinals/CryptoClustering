##### Import dependencies and libraries

```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

##### Load the csv file and display the first few rows of the data

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

##### Summary statistics

```python
# Generate summary statistics
df_market_data.describe()
```

##### Plot the data to visualize market trends

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90,
    xlabel='coin_id',  # Label for the x-axis
    ylabel='Values',   # Label for the y-axis
    title='Market Data Trends'  # Add a title to the plot
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
normalized_data = scaler.fit_transform(df_market_data)
normalized_data[:5]
```

##### Create a dataframe with scaled data

```python
# Create a DataFrame with the scaled data
df_scaled_data = pd.DataFrame(normalized_data, columns=df_market_data.columns, index=df_market_data.index)

# Copy the crypto names from the original data
df_scaled_data['coin_id'] = df_market_data.index

# Set the coinid column as index
df_scaled_data.set_index('coin_id', inplace=True)

# Display sample data
df_scaled_data.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
k_values
```

##### Compute inertia for each k-value

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # Inside the loop:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, random_state=0)
    # 2. Fit the model to the data using `df_market_data_scaled`
    model.fit(df_scaled_data)
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)
```

##### Create a dictionary with k-values and inertia
##### Convert the dictionary to a dataframe

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': k_values,      # List of k-values
    'inertia': inertia  # Corresponding inertia values
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

##### Plot the elbow curve

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title='Elbow Curve for Optimal k (Original Data)',
    xticks=k, 
    xlabel='Number of Clusters (k)', 
    ylabel='Inertia', 
    width=800, 
    height=400
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Based on the Elbow Curve, the best value for k is 4.

- The "elbow point" is where the inertia starts to decrease at a slower rate, which indicates diminishing returns from adding more clusters.
- In this case, the slope of the curve flattens significantly after k=4, suggesting that 4 is the optimal number of clusters for this data.

### Cluster Cryptocurrencies with K-means Using the Original Data

##### Initialize kmeans model

```python
# Initialize the K-Means model using the best value for k
k = 4  # Based on the Elbow method
kmeans_model = KMeans(n_clusters=k, random_state=0)
```

##### Fit the KMeans model to the scaled data to learn the clusters.

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(df_scaled_data)
```

##### Predict and print the cluster labels for each cryptocurrency using the KMeans model.

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
cluster_predictions = kmeans_model.predict(df_scaled_data)

# Print the resulting array of cluster values.
print("Cluster labels for each cryptocurrency:")
cluster_predictions
```

##### Copy of original dataframe

```python
# Create a copy of the DataFrame
df_crypto_clusters = df_market_data.copy()
```

##### Added new column with predicted clusters

```python
# Add a new column to the DataFrame with the predicted clusters
df_crypto_clusters['Cluster'] = cluster_predictions

# Display sample data
df_crypto_clusters.head()
```

##### Scatter plot of cryptocurrency clusters based on 24h vs 7d price changes, colored by KMeans cluster labels with hover information showing crypto names and price changes.

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
scatter_plot = df_crypto_clusters.hvplot.scatter(
    x='price_change_percentage_24h',         # X-axis feature
    y='price_change_percentage_7d',          # Y-axis feature
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.          
    c='Cluster',                             
    hover_cols=['coin_id'],
    cmap='Category10',                       # Use a darker, more distinct color map                  
    title='Cryptocurrency Clusters',         # Add a title
    xlabel='24h Price Change (%)',           # Label for the X-axis
    ylabel='7d Price Change (%)',            # Label for the Y-axis
    width=800,                               # Plot width
    height=600                               # Plot height
)

# Display the scatter plot
scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

##### Create a PCA model to reduce the data to 3 principal components.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

##### Reduce the data to 3 principal components and show the first 5 rows.

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_data = pca.fit_transform(df_scaled_data)

# View the first five rows of the DataFrame.
pca_data[:5]
```

##### Print the explained variance for each principal component and the total.

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

# Print the explained variance ratio for each of the three principal components
print("Explained Variance Ratio for each principal component:")
print(f"PC1: {explained_variance[0]:.4f}")
print(f"PC2: {explained_variance[1]:.4f}")
print(f"PC3: {explained_variance[2]:.4f}")

# Calculate and print the total explained variance of the three components
total_explained_variance = explained_variance.sum()
print(f"Total Explained Variance (PC1 + PC2 + PC3): {total_explained_variance:.4f}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is 0.8950, which means that these three components together explain 89.50% of the variability in the dataset.

##### Create PCA DataFrame, set coin_id as index, and display data

```python
# Create a new DataFrame with the PCA data.
# Creating a DataFrame with the PCA data
df_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original data
df_pca['coin_id'] = df_market_data.index

# Set the coinid column as index
df_pca.set_index('coin_id', inplace=True)

# Display sample data
df_pca.head()
```

### Find the Best Value for k Using the PCA Data

##### List of k-values for clustering

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

##### Compute and store inertia for each k using PCA data.

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # Inside the loop:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, random_state=0)
    # 2. Fit the model to the data using `df_market_data_pca`
    model.fit(df_pca)
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)
```

##### Create a dictionary and DataFrame to plot the Elbow curve.

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': k_values,      # List of k-values
    'inertia': inertia  # Corresponding inertia values
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the first few rows of the DataFrame
df_elbow.head()
```

##### Plot the Elbow curve to visually find the optimal k

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title='Elbow Curve for Optimal k (PCA)',
    xticks=k_values, 
    xlabel='Number of Clusters (k)', 
    ylabel='Inertia', 
    width=800, 
    height=400
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k appears to be 4, as this is where the curve begins to flatten, indicating diminishing returns in reducing the inertia with additional clusters.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, it does not differ from the best k value found using the original data, which was also 4.
